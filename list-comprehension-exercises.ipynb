```python
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

df = sns.load_dataset("car_crashes")
```

Using the List Comprehension structure, convert the names of the numeric variables in the car_crashes data to uppercase and add NUM in front of them. 
Non-numeric ones should also be converted.
It should be done with a single list comp structure.

```python
num_cols_upper = ["NUM_" + col.upper() if df[col].dtypes in [int, float] else col.upper() for col in df.columns]
print(num_cols_upper)
```

Using the List Comprehension structure, put "FLAG" at the end of the names of variables in the car_crashes data that do not contain "no".

Notes:
All variable names must be in uppercase.
It should be done with a single list comp.

```python
col_flag = [col.upper() + '_FLAG' if "NO" not in col.upper() else col.upper() for col in df.columns ]
print(col_flag)
```

Using the List Comprehension structure, select the names of the variables that are DIFFERENT from the variable names given below and create a new dataframe.
og_list = ["abbrev", "no_previous"]

Notes:
First, create a new list named new_cols using list comprehension according to the list above.
Then create a new df by selecting these variables with df[new_cols] and name it new_df.

```python
og_list = ["abbrev", "no_previous"]
new_cols = [col for col in df.columns if col not in og_list]
new_df = df[new_cols]
print(new_df.head(5))
```
